# Creating Dataframes

We create dataframes, viewing the basic statistics for each dataset.

```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML

def style(html):
    return f"""
    <style>
      .styled-table {{
        border-collapse: collapse;
        font-size: 13px;
        font-family: monospace;
        width: 100%;
        white-space: nowrap;
      }}
      .styled-table thead th {{
        background-color: #f0f0f0;
        position: sticky;
        top: 0;
        z-index: 2;
        box-shadow: 0 1px 0 #ddd;
      }}
      .styled-table th, .styled-table td {{
        border: 1px solid #ddd;
        padding: 6px 12px;
        text-align: left;
      }}
      .styled-table tbody tr:nth-child(even) {{
        background-color: #fafafa;
      }}
      .styled-table tbody tr:hover {{
        background-color: #eef4ff;
      }}
    </style>
    <div style="overflow-x: auto; border: 1px solid #ddd; border-radius: 6px;">
      {html}
    </div>
    """


def print_statistics(df: pd.DataFrame):
  print(f'Number of rows: {len(df)}')
  print(f'Number of columns: {len(df.columns)}')
  print("Columns: ", end = ' ')
  print(', '.join(df.columns))
  print('Dataset Sample: \n')
  sample = df.sample(5)
  styled_html = style(sample.to_html(index=True))
  display(HTML(styled_html))
```

```python
df_steam = pd.read_csv("./dataset/steam.csv")
df_reqs = pd.read_csv("./dataset/steam_requirements_data.csv")
df_sup = pd.read_csv("./dataset/steam_support_info.csv")
df_tags = pd.read_csv("./dataset/steamspy_tag_data.csv")
```

```python
# Game statistics
print_statistics(df_steam)
```

```python
# Game requirement statistics
print_statistics(df_reqs)
```

```python
# Game support statistics
print_statistics(df_sup)
```

```python
# Game tag statistics
print_statistics(df_tags)
```
